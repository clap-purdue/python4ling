## Varibale Names:
- All varibale names are valid except starting with a numerical value or containing `.`. Both `3varibale` and `this.varibale` are INVALID.
- Variable names cannot contain one of the words reserved by python:
    - `for`, `while`, `break`, `pass`, `continue`, `try`, `except`
    - `in`, `not`, `is`, `import`, `as`, `with`, `if`, `from`, `else` 
    - `def`, `class`, `return`, `yield`, `raises`, `elif`, `finally`

### Practice:
- valid vs. invalid variable name:
    - `asd`
    - `as12`
    - `for`
    - `123`
    - `None`
    - `12_asd`
    - `num_py`

```python
import re
def check(name):
    python_reserved_word = ["for", "while", "break", "pass", "continue", "try", "except", "in", "not", "is", "import", "as", 
                            "with", "if", "from", "else", "def", "class", "return", "yiled", "raises", "elif", "finally", "None"]
    if name in  python_reserved_word:
        return "Invalid"
    if not isinstance(name, str):
        return "Invalid"
    if not re.match(r'^[a-zA-Z_][a-zA-Z0-9_]*$', name):
        return "Invalid"
    return f"`{name}` is a valid varibale name."
```

```python
print(check("asd"))
print(check("as12"))
print(check("for"))
print(check("123"))
print(check("None"))
print(check("12_asd"))
print(check("course"))
```

- In Python, the term “object” is very important. 

- Objects include numbers, strings of characters, lists, functions.

# What is an OBJECT in python?

*NOTE:* this might be advanced. If we could not get the whole idea, no worries. We will come to this later.

Before we start, let's understand two things:

1. What is an object in python?
2. What are methods (built-in functions)?

```python
# 1. What is an object in python?
# This is an object: a very simple calculator

class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
        
    def add(self):
        return self.num1 + self.num2
```

```python
# an object is anything that can be assigned to a variable
calculator = Calculator(10, 20)
#print(calculator.add())
print(type(calculator))
# we use the method `add` to add the two numbers
#print(calculator.add())

# lets' check the type
#print(type(calculator))

# lets' use isinstance
#print(isinstance(calculator, Calculator))
```

```python
# 2. What are methods?

# this is a function but `add()` in the previous example is a method
def calc(num1, num2, operation):
    if operation == "add":
        return num1 + num2
    elif operation == "subtract":
        return num1 - num2
    elif operation == "multiply":
        return num1 * num2
    elif operation == "divide":
        return num1 / num2
  
```

```python
calc(1,2, "add")
```

## Operations on Strings:

### Indexing and Slicing

- Indexing means accessing individual items from a sequence by specifying the index of that item.

| P | Y | T | H | O | N |
| :-|:-:|:-:|:-:|:-:|-: |           
| 0 | 1 | 2 | 3 | 4 | 5 | 
| -6 | -5 | -4 | -3 | -2 | -1 |

- we have `positive` and `negative` indexes.

- strings are sequences of characters and can be sliced.

- Slicing syntax is as follows `seq[start, stop, step]`


```python
# get index 0 which will be the P letter - positive
word = "PYTHON"
word[4]
```

```python
# get index 3 which will be the H letter - positive
word = "PYTHON"
word[3]
```

```python
# get index -1 which will be the N letter - negative
word = "PYTHON"
word[3:6]
```

```python
# slicing with step
word = "PYTHON"
word[0:4:1]
```

```python
# get all
word = "PYTHON"
word[:]
```

```python
# get what's after index 2 - note that index 2 is included.
word = "PYTHON"
word[2:]
```

```python
# get what's before index 2 - note that index 2 is NOT included.
word = "PYTHON"
word[:2]
```

```python
# reverse the order using negaive step
word = "PYTHON"
word[::-1]
```

```python
# use negative to get the last 3 letters
word = "PYTHON"
word[-3:]
```

```python
# using the `slice()` built-in function
# word = "PYTHON"
# word[slice(0,2,1)]
```
