### `append(elmnt)`

- appends an element to the end of the list. It takes one `required` argumant.

```python
colors = ['red', 'blue', 'yellow']
colors.append("black")
colors
```

```python
# we can also append a list
l1 = ['red', 'blue', 'yellow']
l2 = ['black', 'brown', 'green']
l1.append(l2)
l1
```

### `clear()`

- The clear() method removes all the elements from a list.

```python
colors = ["red", "blue"]
colors.clear()
colors
```

### `pop(pos)`

- removes the element at the specified position. It takes one `required` argument which is a number that specifies the position.


```python
colors = ["red", "blue", "green"]
colors.pop(0)
colors
```

### `remove(elmnt)`

- removes the first occurrence of the element with the specified value. It takes one `required`.


```python
colors = ["red", "blue", "yellow"]
colors.remove("blue")
colors
```

### `copy()`

- returns a copy of the specified list.

```python
colors = ["red", "blue", "yellow"]
x = colors.copy()
x
```

### `count(value)`

- returns the number of elements with the a value. It takes one `required` argument.


```python
colors = ["red", "blue", "yellow", "red", "yellow"]
x = colors.count("red")
x
```

### `extend(iterable)`

- adds the elements of any iterable to the end of a list. It takes one `required` argument (iterable).

```python
colors = ["red", "blue", "red", "yellow"]
num = (1, 4, 5, 9)

colors.extend(num)
colors
```

### `index(elmnt)`

- returns the position at the first occurrence of the specified value. It takes one `required` argument.

```python
nums = [43, 3, 76, 23, 12]
x = nums.index(76)
x
```

### `insert(pos, elmnt)`

- inserts the specified value at the specified position. It takes 2 `required` arguments. `pos` is a number and `elmnt` is an element to  be inserted.

```python
vowels = ['a', 'e', 'o', 'u']
vowels.insert(2, 'i')
vowels
```

### `reverse()`

- reverses the sorting order of the elements.

```python
w = ['c', 'a', 'r']
w.reverse()
w
```

### `sort(reverse, key)`

- sorts the list ascending by default. It takes 2 `optional` argumants.

```python
vowels = ['a', 'e', 'i', 'o', 'u']
vowels.sort()
vowels
```

```python
vowels = ['a', 'e', 'i', 'o', 'u']
vowels.sort(reverse=True)
vowels
```
