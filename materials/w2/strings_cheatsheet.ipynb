### `split(separator, maxsplit)`

- It splits a string into a list of strings. It can take two `arguments`:
  - `separator` is optional and it is the separator used tp split the string.
  - `maxsplit` is optional and it determines the number of split.

```python
# the default separator is white spaces
text1 = "welcome to Purdue."

res = text1.split()

print(res) 
```

```python
# use separator and max_splits
text2 = "hello, my name is Python, and I am a wonderful programming language."

res = text2.split(", ", 1)

print(res) 
```

### `rsplit(separator, maxsplit)`

- It splits a string into a list of strings. It starts from the `right`. It can take two `arguments` of `separator` and `maxsplit` as `split()`

```python
res = text2.rsplit(", ")

print(res) 
```

```python
res = text2.rsplit(", ", 2)

print(res) 
```

### `splitlines(separator, maxsplit)`

- This method splits a string into a list at line breaks. It can have `keeplinebreaks` as an optional argument to specify if we include the linebreak or not.

```python
# `False` is the default. 
text3 = "hello, my name is Python\nI am a wonderful programming language."

res = text3.splitlines()

print(res)
```

```python
# `True` will show the linebreak
text3 = "hello, my name is Python\nI am a wonderful programming language."

res = text3.splitlines(True)

print(res)
```

### `strip(characters)`

- It removes any whitespaces at the beginning or end of a string. It can take `characters` as optional argument if we want to remove a set of characters.

```python
text4 = "     python     "

print(f"{text4} is awesome.")

res = text4.strip()

print(f"{res} is awesome.")
```

```python
text5 = "xxxx...,f!!!Python...ddd"

res = text5.strip(".,xf!d")

print(f"{res} is awesome.")
```

### `rstrip(characters)`

- It removes any whitespaces at the end of a string. It can take `characters` as optional argument if we want to remove a set of characters.

```python
res = text4.rstrip()
res
```

```python
res = text5.rstrip(".,xf!d")
res
```

### `upper()`

### `join(iterable)`

- it takes all elements and joins them into one string. It `must` have one argument which is the `iterable` meaning a list, tuple, et cetera.

```python
names = ("Marley", "Claire", "Dominic")

res = "#".join(names)

print(res)
```

### `format(value1, value2, ...)`

- It formats the specified value(s) and insert them inside the string's placeholder.

- The placeholder is defined using curly brackets: {}. 

```python
# placeholders
# 1) 
text10 = "My name is {name}, I'm {age}".format(name = "Abby", age = 20)
print(text10)

# 2)
text11 = "My name is {0}, I'm {1}".format("Abby",20)
print(text11)

# 3) 
text12 = "My name is {}, I'm {}".format("Abby",20)
print(text12)
```

```python
# formatting styles
# (1) `:<` means Left aligns the result (within the available space)
text13 = "I am {:<8} years old."
print(text13.format(20))

# read more on formatting styles: https://docs.python.org/3/library/string.html#formatspec
```

```python
# using f string. Here you do not need to use format()
name = "Abby"
age = 20

text14 = f"My name is {name}, I'm {age}"
text14
```

### `replace(oldvalue, newvalue, count)`

- This method replaces a specified string with another. It takes three `arguments`:
   - `oldvalue`(required): the old string.
   - `newvalue`(required): the new string.
   - `count`(optional): specify how many occurences of the old value you want to replace.


```python
text15 = "I like Java"

res = text15.replace("Java", "Python")
print(res) 
```

### `count(value, start, end)`

- It returns the number of times a certain value appears in the string.
- It takes 3 arguments:
    - `value`(required): the value we are looking for.
    - `satrt`(optional): the position to start the search. Default is 0
    - `end`(optional): the position to end the search. Default is the end of the string

```python
text15 = "Python is amazing. I love Python."

res = text15.count("Python")
print(res)
```

```python
res = text15.count("Python", 0, 40)
print(res)
```

### `find(value, start, end)`

- It finds the first occurrence of the specified value.

- It returns -1 if the value is not found. It is similar to `index()`

- It has 3 arguments:
  - `value`(required): the value we are looking for.
  - `start`(optional): the start of the search. Default is 0.
  - `end`(optional): the end of the search. Default is the end of the string

```python
text16 = "welcome to Purdue."

res = text16.find("to")
print(res)
```

```python
# find() and index()
print(text16.find("t"))
print(text16.index("t"))
```

### `startswith(value, start, end)`

- returns True if the string starts with the specified value, otherwise False.

- It has 3 arguments:
   - `value`(required)
   - `start`(optional)
   - `end`(optional)

```python
str1 = "Hello, welcome to Purdue."

res1 = str1.startswith("Hello")
print(res1) 

res2 = str1.startswith("llo", 2, 50)
print(res2) 
```

### `endswith(value, start, end)`

- returns True if the string ends with the specified value, otherwise False.

- It has 3 arguments:
   - `value`(required)

   - `end`(optional)

```python
res3 = str1.endswith("due.")
print(res3) 

res4 = str1.endswith("e.", 3, 50)
print(res4) 
```

### `center(length, character)`

- aligns the string to the center, using a specified character.
- space is default as the fill character. It takes 2 parameters: `length` (required) and `character` (optional).


```python
other1 = "python"

res = other1.center(20)

print(res)
```

```python
other2 = "python"

res = other2.center(20, "X")

print(res)
```

### `encode(encoding, errors)`

- encodes a string using the specified encoding. 
- `utf-8` is the deafult encoding.
- It takes two arguments: `encoding` (optional) and `errors` (optional).


```python
other3 = "اسمي السّيّد"

res = other3.encode()

print(res)
```

```python
other4 = "اسمي السّيّد"

res = other3.encode(encoding="ascii", errors="backslashreplace")

print(res)
```

### `expandtabs(tabsize)`

- adjusts the tab size to the specified number of whitespaces.
- It takes 1 argument: `tabsize`(optional). Default is 8

```python
other5 = "P\ty\tt\th\to\tn"

res =  other5.expandtabs()

print(res)
```

### `index(value, start, end)`

- finds the first occurrence of a value.

- It takes three qrguments: `value`(required), `start`(optional), `end`(optional).

```python
other6 = "Python"

res = other6.index("P")

print(res)
```

### `ljust(length, character)`

- It aligns the string to the left using a character (space is default).

- It takes 2 arguments: `length` (required) and `character` (optional).

```python
other7 = "Python"

res = other7.ljust(20)

print(res, "is my favorite programming language.")
```

```python
other8 = "Python"

res = other8.ljust(10, "A")

print(res)
```
