# Global Variables

- In functional programming, avoiding global variables is generally recommended for several practical reasons such as readability, unexpected mutation, and concurrency issues.

````python
gv = 0                    #create global variable
def f1(x):                #function to manipulate global
    global gv
    gv += x

print(gv)                #print initial value of global
f1(3)                   #invoke the function
print(gv)               #print the new global value
````

- In this code, as we trace the flow of the program, we must keep track of not just what the function might do and the variables it binds, but the state of global variables that exist outside the current function.

````python
gv = 0                     #create global variable
def f1(g,x):               #function to manipulate global 
    g += x
    return g

print(gv)                   #print initial value of global 
gv = f1(3,gv)               #invoke the function
print(gv)                    #print the new global value
````

# `lambda` function

- Lambda functions are just small, anonymous (unnamed) functions written in a single line.
```python
lambda arguments: expression
```

##### Simple comparison
- Both do the exact same thing. 
- The lambda version just skips the def, the name, and the return.
- The expression is automatically returned.

````python
def double(x):                   # Regular function
    return x * 2

print(double(5))  
````

````python
double_lam = lambda x: x * 2        # Same thing as a lambda

print(double_lam(5))
````

##### Multiple Arguments work too


````python
add = lambda a, b: a + b
print(add(3, 4))
````

##### Sorting with key

````python
names = ["Charlie", "Alice", "Bob"]
def get_length(name):
    return len(name)

names.sort(key=get_length)
names
````

````python
names2 = ["Charlie", "Alice", "Bob"]

names2.sort(key=lambda name: len(name))
print(names2)
````

##### Sorting dictionaries

````python
people = [
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age": 35},
]

# by age
people.sort(key=lambda person: person["age"])
people
````

##### With `map` and `filter`

````python
numbers = [1, 2, 3, 4, 5]

# double every number
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)  

# keep only even numbers
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens) 
````

#### `map()`
- The `map()` function applies some function to every element in a sequence, returning the output for each application in a list.

````python
x = ['Tom','Dick','Harry']    #a list of strings

y = map(len,x)                #create a map object where len() #is applied to each string

print(y)                      #print map object (not too useful)
print(list(y))                #convert map object to a list 
y = map(len,x)                #recreate map object
for i in y:                   #iterate through map object
    print(i)
````

#### `zip()`
- The `zip()` function takes a set of sequences and constructs a sequence of tuples, drawing one element from each sequence in turn. 
- It returns an object that can be iterated over or that can be converted into a list. 

````python
x = zip('cat','dog')      #make pairs object 
print(list(x))            #convert to list

x = zip('cat','dog')     #make pairs again
for p in x:              #iterate through pairs
    print(p)
````

#### `filter()`
- The `filter()` function is rather similar; it applies a boolean test to every element in a sequence and returns only those elements for which the test returns True.

````python
a = ['Tom','Dick','Harry','Mike','Abby'] #a list of strings

b = filter(lambda x: len(x) == 4,a)      #create a filter object where a lambda 
                                         #function is applied to each element
 
print(list(b))                           #convert map object to a list
````

#### `reduce()`
- The `reduce()` function is from the `functools` module. 
- This function allows you to reduce a sequence of elements to one element by applying a function that takes two arguments. 

````python
from functools import reduce

a = list(range(10))               #create list of numbers 
print(a)

b = reduce(lambda x,y: x+y,a)     #reduce with addition
print(b)                          #print result
````

````python
from functools import reduce

a = list(range(5))                  #create list of numbers
print(a)

b = reduce(lambda x,y: x+y,a,9)     #reduce with addition
print(b)  

c = reduce(lambda x,y: x+y,[],9)    #another reudction - nothing to redcue, just empty list
print(c) 
````

## Overt Recursion
- Recursion: functions that are defined with respect to themselves.
- `5! = 5 x 4 x 3 x 2 x 1`

````python
def fac(n):
    if n == 1:                    #base case of recursion
        return 1
    else:                         #recursive clause
        return (n * fac(n-1))     #invokes the function ITSELF

print('1! =',fac(1))           #invoked with base case 
print('5! =',fac(5))           #invoked with recursive case 
````

````python
import math
print(math.factorial(5))
````

- Another example

````python
def pfx1(s):                    #recursive function to find prefixes
    if len(s) > 0:
        print(s)
        pfx1(s[:-1])
        
def pfx2(s): 
    """
    range(len(s),0,-1)): generates numbers from len of s down to 1.
    s[:x]: slices string from the beginning up to x.
    list() forces map() to apply.
    """
    list(map(lambda x: print(s[:x]),range(len(s),0,-1)))

pfx1('happy')  
print()
pfx2('happy')
````

## Comprehensions
- Comprehensions are powerful tools in Python. 
- They allow us to create lists, sets, or dictionaries from other sequences.
- They enable us to create various sequences while avoiding overt control structures. 
- The basic syntax for a list comprehension is:
`[expression for item in iterable]`


### Lists:

````python
# Given this list of integers [0,1,2,3,4], let's create a list of their squares

list_of_ints = [0,1,2,3,4]
squares = []

for x in list_of_ints:
    squares.append(x**2)
    
print(squares)
````

````python
squares = [x**2 for x in list_of_ints]
print(squares)
````

## Vectorized Computation:

- Second-order functions, recursion, and comprehensions can all be used to avoid control structures and move us toward the functional programming ideal. 
- Another way to do this is vectorized computation or vectorization. 
- Computation is vectorized when operations on a sequence of elements can be performed in a single step.

````python
a = [1,2,3]
b = [4,5,6]
print(a+b)

c = []
for i in range(0,len(a)):    #go through old lists
    c.append(a[i]+b[i])     #add current values together 
print(c) 
````

- One way to avoid this is to use zip() and map():

````python
a = [1,2,3] 
b = [4,5,6]
c = zip(a,b) 
d = list(map(sum,c))
print(d) 
````

- The `numpy` module provides this same functionality for all mathematical operations. 
- What we do is convert our lists to numpy arrays, and then mathematical operators have their vectorized effects. 

````python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

c = a + b
d = a * b
print(c)
print(d)
````

## Iterables, Iterators, and Generators
- We look more closely at iteration. 
- We start with `iterables`. 
- These are data structures that can be iterated over with control structures like for or while and include lists, dictionary keys, strings, and files. 

````python
# lists
a = [1,2,3,4,5]
for i in a:
    print(i+1)

# dictionaries    
b = {'Tom':14,'Dick':25,'Harry':8}          
for k, v in b.items():
    print(k,v)
````

- An `iterator` is a special kind of iterable. 
- They come in two forms. 
- 1. One form uses OO programming and is based on the `__next__()` method:

````python
#an iterator that represents a sequence
#of integers
class Sequence:
    #initialize endpoints
    def __init__(self,low,high):
        self.now = low
        self.high = high
    #required for iterators
    def __iter__(self):
        return self
    #gets the next item
    def __next__(self):
        #must throw this error at limit
        if self.now > self.high:
            raise StopIteration
        #return current value and increment
        else:
            i = self.now
            self.now += 1
            return i

s = Sequence(3,10)                    #create a new sequence

for i in s:                           #iterate through the sequence
    print(i)
````

- 2. The other kind of iterator is built around the `__getitem__()` method. 
- The basic idea here is to build a list that we can index into.

````python
#a different kind of iterator
class Students:
    #initialize with list of names
    def __init__(self,listofstudents):
        self.students = listofstudents
    #how to change a name
    def __setitem__(self,n,name):
        self.students[n] = name
    #how to retrieve names
    def __getitem__(self,n):
        return self.students[n]
    #returns number of names
    def __len__(self):
        return len(self.students)
    #adds a name to the list
    def append(self,name):
        self.students.append(name)
    #deletes an item
    def __delitem__(self,n):
        del self.students[n]

#create a new student list
s = Students(['Mike','Joey'])
s[1] = 'Diane'              #change name of 2nd student
s.append('Bob')             #add a student
del s[0]                    #delete a student
for n in s:                 #iterate through final list
    print(n)
````
