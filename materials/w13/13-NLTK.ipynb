### Lexical Diversity (LD)

- Lexical diversity measures the lexical sophistication used in a given text. It quantifies the variety of words used in a text in relation to its length. 

- There are several metrics used to measure lexical diversity. 

- `Types:` The number of unique words (or word forms) in a text.
- `Tokens:` The total number of words in a text.

#### Interpretation
- `Higher scores`: indicate greater lexical diversity. This suggests that the text uses a wide range of different words and potentially has a rich vocabulary.

- `Lower scores`: indicates less lexical diversity. This could suggest that the text uses a more limited vocabulary, with repetition of the same words.

#### Applications:

- `Comparing authors`
- `Language development`
- `Identifying writing style` 


### Type-Token Ratio (TTR)


$$ TTR = {types \over tokens} $$




```python
l = "Lexical diversity measures the lexical sophistication used in a given text. It quantifies the variety of words used in a text in relation to its length."


types = set(l.split())
tokens = l.split()

ttr = len(types)/ len(tokens)
ttr
```

### Root Type-Token Ratio (RTTR)

$$ RTTR = {types \over \sqrt{tokens}} $$

### Corrected Type-Token Ratio (CTTR)

$$ CTTR = {types \over \sqrt{2 * tokens}} $$

### Log Type-Token Ratio (LogTTR)

$$ LogTTR = {log_{10}{types} \over log_{10}{tokens}} $$

### Mass

$$ Mass_{a^2} = {{log_{10}{types} - log_{10}{tokens}} \over (log_{10}{tokens})^2} $$

### Mass

$$ a = {\sqrt Mass_{a^2}} $$

- There are many metrics such as `Uber`, `Yules's K`, `Mean Segmental Type Token Ratio (MSTTR)`, `Moving Average Type Token Ratio (MATTR)`, `vocd-D`, `HD-D (hypergeometric probability)-D`, `Measure of Textual Lexical Diversity (MLTD)`, et cetera.

- If you interested, you can read about them and discover how they work.

```python
import nltk
alice = nltk.Text(gutenberg.words("carroll-alice.txt"))
alice.concordance("rabbit")
```

```python
# Find length of each word in alice
alice_word_length = []

for word in alice:
    alice_word_length.append(len(word))

#alice_word_length
```

```python
# find the longest words in alice
alice_word_length = [len(word) for word in alice]
max_length = max(alice_word_length)

l = []
for word in alice:
    if len(word) == max_length:
        l.append(word)

l
```

```python
# Some stats about texts in GutenBerg
# This computes:
# 1) average word length, 
# 2) average sentence length, and 
# 3) the number of times each vocabulary item appears in the text on average.
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
```

### 3. Brown Corpus

- The Brown Corpus was the first million-word electronic corpus of English, created in 1961 at Brown University.

- The Brown Corpus is a convenient resource for studying systematic differences between genres, a kind of linguistic inquiry known as `stylistics`.

```python
from nltk.corpus import brown
brown.categories()
```

```python
brown_raw = brown.raw(categories='adventure')
len(brown_raw)
```

```python
brown_words = brown.words(categories='news')
len(brown_words)
```

```python
brown_words = brown.words(fileids=['cg22'])
```

```python
brown_sent = brown.sents(categories=['news', 'editorial', 'reviews'])
len(brown_sent)
```

```python
# An example of studying stylistics
from nltk.corpus import brown

news_text = brown.words(categories='news')

fdist = nltk.FreqDist(w.lower() for w in news_text)
modals = ['can', 'could', 'may', 'might', 'must', 'will']
for m in modals:
    print(m + ':', fdist[m], end=' ')
```

```python
cfd = nltk.ConditionalFreqDist((genre, word) for genre in brown.categories() for word in brown.words(categories=genre))
genres = ['news', 'religion', 'hobbies', 'science_fiction', 'romance', 'humor']
modals = ['can', 'could', 'may', 'might', 'must', 'will']
cfd.tabulate(conditions=genres, samples=modals)
```

```python
# cumulative frequency plot of word lengths across six different languages
# The x-axis represents word length (1 letter, 2 letters, etc. up to 37), 
# and the y-axis shows the cumulative count of words at that length or shorter. 
languages = ['Chickasaw', 'English', 'German_Deutsch','Greenlandic_Inuktikut', 'Hungarian_Magyar', 'Ibibio_Efik']
cfd = nltk.ConditionalFreqDist(
    (lang, len(word))
    for lang in languages
    for word in udhr.words(lang + '-Latin1'))

cfd.plot(cumulative=True)
```
