### What is a virtual environment?
- A virtual environment is like a separate, isolated box for each of your Python projects. 
- Each box has its own copy of Python and its own set of installed packages. 
- This way, Project A can use pandas 1.5 while Project B uses pandas 2.0 — and they won't interfere with each other.

## Option 1: Using Python's Built-In `venv`

`venv` comes with Python 3 — no extra installation needed.

### Step 1: Create a Virtual Environment

Open your terminal and navigate to your project folder:

```bash
cd my_project
```

Then create the environment:

```bash
python -m venv myenv
```

This creates a folder called `myenv` inside your project directory. You can name it anything, but `myenv`, `venv`, and `.venv` are common choices.

### Step 2: Activate It

**Mac/Linux:**

```bash
source myenv/bin/activate
```

**Windows (Command Prompt):**

```cmd
myenv\Scripts\activate
```

**Windows (PowerShell):**

```powershell
myenv\Scripts\Activate.ps1
```

Once activated, you will see `(myenv)` at the beginning of your terminal prompt. This tells you the environment is active.

### Step 3: Install Packages

Now any packages you install go into this environment only:

```bash
pip install numpy pandas matplotlib
```

### Step 4: Save Your Dependencies

To record what you've installed (so others can recreate your environment):

```bash
pip freeze > requirements.txt
```

Someone else can then install the same packages with:

```bash
pip install -r requirements.txt
```

### Step 5: Deactivate When Done

```bash
deactivate
```

### Step 6: Delete an Environment

Just delete the folder:

```bash
rm -rf myenv        # Mac/Linux
rmdir /s myenv      # Windows
```

## Option 2: Using Anaconda (conda)

Anaconda is popular in data science because it manages both Python packages and non-Python dependencies.

### Step 1: Create a Conda Environment

```bash
conda create --name myenv python=3.11 ipython
```

This creates an environment called `myenv` with Python 3.11. You can pick any Python version you need.

### Step 2: Activate It

```bash
conda activate myenv
```

You'll see `(myenv)` in your prompt.

### Step 3: Install Packages

You can use either `conda install` or `pip install`:

```bash
conda install numpy pandas matplotlib
```

Prefer `conda install` when the package is available. It handles dependencies better. Fall back to `pip install` for packages not in the conda repositories.

### Step 4: See What's Installed

```bash
conda list
```

### Step 5: Save Your Environment

```bash
pip freeze > requirements.txt
```


```bash
conda env export > environment.yml
```

Someone else can recreate it with:

```bash
conda env create -f environment.yml
```

### Step 6: Deactivate

```bash
conda deactivate
```

### Step 7: List All Your Environments

```bash
conda env list
```

### Step 8: Delete an Environment

```bash
conda env remove --name myenv
```

- For more, see conda cheat sheet: https://docs.conda.io/projects/conda/en/4.6.0/_downloads/52a95608c49671267e40c689e0bc00ca/conda-cheatsheet.pdf

## Using Virtual Environments in Jupyter Notebooks

By default, Jupyter might only see your base Python — not your virtual environments. Here's how to fix that.

### Method A: venv + Jupyter

1. **Activate your environment**.

2. **Install Jupyter and the IPython kernel package:**

```bash
pip install jupyter ipykernel
```

3. **Register the environment as a Jupyter kernel:**

```bash
python -m ipykernel install --user --name=myenv
```

- `--name` is the virtual env name

4. **Launch Jupyter:**

```bash
jupyter notebook
jupyter lab
```

5. **Select your kernel:** When creating a new notebook, click **New** and choose **"My Project (myenv)"** from the list. For an existing notebook, go to **Kernel > Change Kernel** and select it.

### Method B: Conda + Jupyter

1. **Activate your conda environment:**

```bash
conda activate myenv
```

2. **Install Jupyter and the kernel package:**

```bash
conda install jupyter ipykernel
```

3. **Register the kernel:**

```bash
python -m ipykernel install --user --name=myenv
```

4. **Launch Jupyter:**

```bash
jupyter notebook
jupyter lab
```

5. **Select the kernel** the same way as above.

### How to Verify You're Using the Right Environment

Run this inside a Jupyter notebook cell:

```python
import sys
print(sys.executable)
```

This prints the path to the Python being used. It should point to your virtual environment, not your system Python.

### Removing a Kernel from Jupyter

If you delete an environment but the kernel still shows up in Jupyter:

```bash
jupyter kernelspec list                   # see all kernels
jupyter kernelspec uninstall myenv        # remove a specific one
```
