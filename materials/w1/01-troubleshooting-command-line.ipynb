<h1 align=center>3. Jupyter Notebook</h1>


- There are many ways to write and run Python code, but we will be using `Jupyter Notebooks.`

- `Jupyter Notebooks` allow us to write active runnable code and text.

- For example:

```python
my_name = "John"
print(my_name)
```

- The idea is that the presentation and explanation of Python examples can be interleaved with working editable code examples.

- The software that does this is `Jupyter`. This is a part of the `Anaconda` installation of Python which is `free`.

- Some of our handouts/presentations/homework will be written using `Jupyter`, so that, as I'm presenting basic python concepts, you'll have little code bits you can play with.

### Running Jupyter

- Download the <code>02-august-22.ipynb</code> notebook from the website to your desktop or any other folder.

- Open the terminal application on your machine.

- Using the <code>cd</code> command, switch to <code>~/Desktop</code> or the designated folder.

- Then type <code>jupyter notebook</code>.

- A window should open that includes <code>02-august-22.ipynb</code>. Click on that.

- You can now run/change the code snippets in this notebook.

### Quiting Jupyter

- To exit the <i>Jupyter</i> program, you first close the notebook window. There are then two ways to quit:

- In the *dashboard* window, click <code>Logout</code>. Then, in the terminal window, hit <code>Control-c</code> <b>twice</b>.

- Alternatively, you can just hit the <code>Quit</code> buton in the *dashboard* window.

### What you can do in Jupyter

- There are two kinds of <i>fields</i> in a notebook: <b>code</b> and <b>text</b>. The code fields for us will always be Python. The text fields are formatted using a combination of <i>HTML</i> and <i>markdown</i>.

- Double-clicking in a text field displays the formatting command. Pressing the <i>run</i> button in the menubar shows the formatted text.

- Code fields can be edited directly. Pressing the <i>run</i> button runs the code.
