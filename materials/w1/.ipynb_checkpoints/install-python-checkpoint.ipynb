<h1 align=center>Python Programming - Installation</h1>

# 1. Check if Python is installed

## Mac

- Open your terminal and type:

```bash
python --version
```

## Windows

- Open Powershell and type:
```bash
python --version
```

# 2. Installation options:

- You can choose one of the following two options:
    1. Official Installer `https://www.python.org/`
    2. Anaconda `https://www.anaconda.com/`


<h1 align=center>Official Installer</h1>

### Official Installer

1. Go to the download page here: `https://www.python.org/downloads/`
2. Download the latest version
3. When the download is finished, run the installer by double-clicking the downloaded file. You should see a window like this one:



4. press continue and it is recommended not to change the destination for installation.
5. After you are done, type `python --version` in your terminal to see the latest python version.
